```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
my_df = pd.read_csv('HTE_dataset_PC_EMC_EC_LiPF6.csv', sep="\t")
my_df
```

```python
my_df = my_df[['T [K]',	'r','c [mol/kg]','k [mS/cm]']].copy(deep=True)
my_df
```

```python
scaler = StandardScaler(with_mean=False)
new_df = scaler.fit_transform(my_df[['T [K]',	'r','c [mol/kg]']])
```

```python
new_df = pd.DataFrame(new_df, columns= ['T [K]',	'r','c [mol/kg]'])
new_df['k [mS/cm]'] = my_df['k [mS/cm]']
```

```python
new_df.to_csv('./standardized_dataset.csv',sep=',', index=False)
```
